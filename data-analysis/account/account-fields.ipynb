```python
import pandas as pd
import json
import requests
```

```python
with open ("/Users/matheusjiran/Adhere/data-analysis/account/account_fields.json") as fp:
    account_req = json.load(fp)
```

```python
fields_json = account_req.get('fields')
```

```python
account_fields = pd.read_json(json.dumps(fields_json))
```

```python
calculated_fields = account_fields.loc[(account_fields['calculated'] == True) & (account_fields['custom'] == True)]
```

```python
# GENERATE INPUTS TO CREATE PROPERTIES
i=0
inputs = []
while i < len(account_fields):
    if (account_fields['calculated'][i] == True and account_fields['custom'][i] == True):
        input = {
        "name": (account_fields['name'][i]).lower(),
        "label": (account_fields['label'][i])+" [SF]",
        "type": "number",
        "fieldType": "calculation_equation",
        "description": account_fields['inlineHelpText'][i],
        "groupName": "sf-calculated",
        "options": [],
        "displayOrder": -1,
        "calculated": True,
        "externalOptions": False,
        "archived": False,
        "hasUniqueValue": False,
        "hidden": False,
        "showCurrencySymbol": True,
        "formField": False,
        "calculationFormula": "closedate - createdate"
        }
        inputs.append(input)
    i+=1

with open("account_fields_inputs.json", 'w') as f:
    json.dump(inputs, f)
```

```python
# GENERATE INPUTS TO UPDATE PROPERTIES

i=0
inputs = []
while i < len(account_fields):
    if (account_fields['calculated'][i] == True and account_fields['custom'][i] == True):
        formula = str(account_fields['calculatedFormula'][i])
        formula = formula.lower()
        input = {
        "name": (account_fields['name'][i]).lower(),
        "label": (account_fields['label'][i]),
        "type": "number",
        "fieldType": "calculation_equation",
        "description": account_fields['inlineHelpText'][i],
        "groupName": "sf-calculated",
        "options": [],
        "displayOrder": -1,
        "calculated": True,
        "externalOptions": False,
        "archived": False,
        "hasUniqueValue": False,
        "hidden": False,
        "showCurrencySymbol": True,
        "formField": False,
        "calculationFormula": "("+formula+")"
        }
        inputs.append(input)
    i+=1

with open("update_account_fields_inputs.json", 'w') as f:
    json.dump(inputs, f)
```

```python
# GENERATE INPUTS TO UPDATE CALCULATED PROPERTIES TO STRING

i=0
inputs = []
while i < len(account_fields):
    if (account_fields['calculated'][i] == True and account_fields['type'][i] == "string"):
        formula = str(account_fields['calculatedFormula'][i])
        formula = formula.lower()
        input = {
        "name": (account_fields['name'][i]).lower(),
        "label": (account_fields['label'][i]),
        "type": "string",
        "fieldType": "calculation_equation",
        "description": account_fields['inlineHelpText'][i],
        "groupName": "sf-calculated",
        "options": [],
        "displayOrder": -1,
        "calculated": True,
        "externalOptions": False,
        "archived": False,
        "hasUniqueValue": False,
        "hidden": False,
        "showCurrencySymbol": True,
        "formField": False,
        "calculationFormula": "if total_commercial__c >= 50000 then 'Yes' else 'No'"
        }
        inputs.append(input)
    i+=1

with open("update_account_fields_inputs.json", 'w') as f:
    json.dump(inputs, f)
```
