```python
with open ("/Users/matheusjiran/Adhere/data-analysis/opportunity/opportunities-fields.json") as fp:
    opp_req = json.load(fp)
```

```python
fields_json = opp_req.get('fields')
```

```python
opp_fields = pd.read_json(json.dumps(fields_json))
```

```python
calculated_fields = opp_fields.loc[(opp_fields['calculated'] == True) & (opp_fields['custom'] == True)]
calculated_fields
```

```python
calculated_fields_names = []
k = 0
while k < len(opp_fields):
    if opp_fields['calculated'][k] == True:
        calculated_fields_names.append(opp_fields['name'][k])
    k+=1
```

```python
with open ("/Users/matheusjiran/Adhere/data-analysis/opportunity/hs-deal-fields.json") as fp:
    hs_fields_req = json.load(fp)

hs_fields = hs_fields_req['results']
hs_fields_df = pd.DataFrame.from_records(hs_fields)
hs_fields_df.to_csv("hs-fields.csv")
```

```python
fields_to_create = []
hs_fields_names = hs_fields_df['name'].tolist()
i=0
for field in calculated_fields_names:
    if field.lower() in hs_fields_names:
        "do nothing"
    else:
        fields_to_create.append(field)
    i+=1
fields_to_create
len(fields_to_create)
```

```python
# GENERATE INPUTS TO CREATE PROPERTIES
i=0
inputs = []
while i < len(opp_fields):
    if (opp_fields['name'][i] in fields_to_create):
        input = {
        "name": (opp_fields['name'][i]).lower(),
        "label": (opp_fields['label'][i]),
        "type": "number",
        "fieldType": "calculation_equation",
        "description": opp_fields['inlineHelpText'][i],
        "groupName": "sf-calculated",
        "options": [],
        "displayOrder": -1,
        "calculated": True,
        "externalOptions": False,
        "archived": False,
        "hasUniqueValue": False,
        "hidden": False,
        "showCurrencySymbol": True,
        "formField": False,
        "calculationFormula": "1 + 1"
        }
        inputs.append(input)
    i+=1

with open("opp_fields_inputs.json", 'w') as f:
    json.dump(inputs, f)

len(inputs)
```

```python
# GENERATE INPUTS TO UPDATE PROPERTIES
i=0
inputs = []
while i < len(opp_fields):
    if (opp_fields['name'][i] in fields_to_create):
        formula = str(opp_fields['calculatedFormula'][i])
        formula = formula.lower()
        input = {
        "name": (opp_fields['name'][i]).lower(),
        "label": (opp_fields['label'][i]),
        "type": "number",
        "fieldType": "calculation_equation",
        "description": opp_fields['inlineHelpText'][i],
        "groupName": "sf-calculated",
        "options": [],
        "displayOrder": -1,
        "calculated": True,
        "externalOptions": False,
        "archived": False,
        "hasUniqueValue": False,
        "hidden": False,
        "showCurrencySymbol": True,
        "formField": False,
        "calculationFormula": "("+formula.replace("probability", "hs_forecast_probability")+")"
        }
        inputs.append(input)
    i+=1

with open("update_opp_fields_inputs.json", 'w') as f:
    json.dump(inputs, f)

len(inputs)
```

```python
# GENERATE HELP TEXT UPDATE
i=0
inputs = []
while i < len(opp_fields):
    formula = str(opp_fields['calculatedFormula'][i])
    formula = formula.lower()
    input = {
    "name": (opp_fields['name'][i]).lower(),
    "label": (opp_fields['label'][i]),
    "description": opp_fields['inlineHelpText'][i],
    }
    inputs.append(input)
    i+=1

with open("description_update_deal_fields.json", 'w') as f:
    json.dump(inputs, f)

len(inputs)
```
