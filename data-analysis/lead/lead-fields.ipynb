```python
import pandas as pd
import json
import requests
```

```python
with open ("/Users/matheusjiran/Adhere/data-analysis/lead/lead-fields.json") as fp:
    lead_req = json.load(fp)
```

```python
fields_json = lead_req.get('fields')
```

```python
lead_fields = pd.read_json(json.dumps(fields_json))
```

```python
# GENERATE INPUTS TO CREATE PROPERTIES
i=0
inputs = []
while i < len(lead_fields):
    if (lead_fields['calculated'][i] == True and lead_fields['custom'][i] == True):
        input = {
        "name": (lead_fields['name'][i]).lower(),
        "label": (lead_fields['label'][i])+" [SF]",
        "type": "number",
        "fieldType": "calculation_equation",
        "description": lead_fields['inlineHelpText'][i],
        "groupName": "sf-calculated",
        "options": [],
        "displayOrder": -1,
        "calculated": True,
        "externalOptions": False,
        "archived": False,
        "hasUniqueValue": False,
        "hidden": False,
        "showCurrencySymbol": True,
        "formField": False,
        "calculationFormula": "1 + 1"
        }
        inputs.append(input)
    i+=1

with open("opp_fields_inputs.json", 'w') as f:
    json.dump(inputs, f)
```

```python
# GENERATE INPUTS TO UPDATE PROPERTIES

i=0
inputs = []
while i < len(lead_fields):
    if (lead_fields['calculated'][i] == True and lead_fields['custom'][i] == True):
        formula = str(lead_fields['calculatedFormula'][i])
        formula = formula.lower()
        input = {
        "name": (lead_fields['name'][i]).lower(),
        "label": (lead_fields['label'][i])+" [SF]",
        "type": "number",
        "fieldType": "calculation_equation",
        "description": lead_fields['inlineHelpText'][i],
        "groupName": "sf-calculated",
        "options": [],
        "displayOrder": -1,
        "calculated": True,
        "externalOptions": False,
        "archived": False,
        "hasUniqueValue": False,
        "hidden": False,
        "showCurrencySymbol": True,
        "formField": False,
        "calculationFormula": "("+formula.replace("probability", "hs_forecast_probability")+")"
        }
        inputs.append(input)
    i+=1

with open("update_opp_fields_inputs.json", 'w') as f:
    json.dump(inputs, f)
```
